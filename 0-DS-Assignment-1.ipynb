You have a matrix $X$ of size (10,5) filled with random values. Print the following :

In this part, you will be working with the library ```pandas```. Import this library with an alias name ```pd```.

````
# Initial imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
````

Download dataset from: http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv
Read the data and display the first 5 rows

````
#Download dataset from: http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv
import io
import requests
file_url = "http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv"
urlresponse = requests.get(file_url).content
pd.set_option('display.max_colwidth', None)
urldata = pd.read_csv(io.StringIO(urlresponse.decode('utf-8')))
urldata.head(5)

````

### Preprocessing of the Dataset



---


**1. Convert the categorical values of columns 'sex', 'embarked', 'pclass' into dummy values and add them into titanic_data_new DataFrame**




---



````
titanic_data_new = pd.get_dummies(urldata,columns=['sex','embarked','pclass'])
titanic_data_new
````



---


**2. Fill the missing data in Age and Fare columns with the mean value of the respected column then add these columns to titanic_data_new**


---



````
titanic_data_new['age'].fillna(titanic_data_new['age'].mean(),inplace=True)
titanic_data_new.loc[titanic_data_new.fare == 0.0000 , 'fare'] = titanic_data_new['fare'].mean()
titanic_data_new.head(17)
````

###Normalization


---



1. Normalize the columns which have numerical data



---



````
normalized_data_age = (titanic_data_new['age']-titanic_data_new['age'].min())/(titanic_data_new['age'].max()-titanic_data_new['age'].min())
normalized_data_fare = (titanic_data_new['fare']-titanic_data_new['fare'].min())/(titanic_data_new['fare'].max()-titanic_data_new['fare'].min())
````



---
2. Drop the original age, fare columns
3. Append the normalized columns to the titanic_data_new DataFrame.



---



````
titanic_data_new['age'] = normalized_data_age
titanic_data_new['fare'] = normalized_data_fare
titanic_data_new.head(22)
````

###Plotting
1. plot the age against the ID(continuous)
2. plot the fare against the ID(scatter)

````
titanic_data_new['age'].plot()
titanic_data_new.reset_index().plot.scatter(x = 'index', y = 'fare')
````

Plot the sine function with domain in $[-3\pi,3\pi]$. Label the axes appropriately. Add title $\sin(x)$. Use latex formatting while adding the title. Set the plot size to (10,10)

````
plt.title('Sin(x)')
plt.xlabel('x axis')
plt.ylabel('y axis')
sin_range = np.arange(-(3*np.pi),(3*np.pi),0.1)
sin_values = np.sin(sin_range)
# plt.size(10,10)
plt.plot(sin_range,sin_values)
plt.axhline()
plt.axvline()
plt.grid()
plt.legend()
plt.show()
````

Plot $\sin(x)$ and its spectrum, $\cos(x)$ and its spectrum in a single plot, in the domain $[-2\pi,2\pi]$. Label the plot appropriately.

````

plt.xlabel('x axis')
plt.ylabel('y axis')
sin_ranges2 = np.arange(-(2*np.pi),(2*np.pi),0.1)
sin_values2 = np.sin(sin_ranges2)
plt.plot(sin_ranges2,sin_values2,label='Sin')
cos_values = np.cos(sin_ranges2)
plt.plot(sin_ranges2,cos_values,label='Cos')
sin_spec = np.fft.fft(sin_values2)
cos_spec = np.fft.fft(cos_values)

plt.plot(sin_ranges2,np.real(sin_spec),label='Sin Spectrum Real Values')
plt.plot(sin_ranges2,np.real(cos_spec),label='Cos Spectrum Real Values')
# plt.plot(sin_ranges2,np.imag(sin_spec),label='Sin Spectrum Imaginary Values')
# plt.plot(sin_ranges2,np.imag(cos_spec),label='Cos Spectrum Imaginary Values')
# plt.gca().set_xlim([-10,10])
plt.legend()
plt.axhline()
plt.axvline()
plt.grid()
plt.show()
````
